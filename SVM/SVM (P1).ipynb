```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(palette='rainbow')
```

### Breask cancer dataset from Scipy

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data.keys()
```

#### checking feature Descripiton

```python
print(data['DESCR'])
```

#### Creating the DataFrame

```python
df_feature = pd.DataFrame(data['data'],columns=data['feature_names'])
```

```python
df_feature.head()
```

```python
df_feature.shape
```

```python
df_feature.info()
```

```python
df_feature.T.head(30)
```

#### Creating Featured and Target/label

```python
featured_x = df_feature
target_y = data['target']
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(featured_x,target_y,test_size=0.2,random_state=123)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

### SVM

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

#### Confusion Matrix and Classification Report

```python
y_preds = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
sns.heatmap(confusion_matrix(y_test,y_preds),annot=True,cbar=False,cmap='summer');
```

```python
print(classification_report(y_test,y_preds))
```

#### GRID SEARCH CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf.get_params()
```

```python
grid_params = {'C':[0.1,1,10,100],
              'gamma':[1,0.1,0.01,0.001]}
```

#### GridSearchCV CLF

```python
grid_clf = GridSearchCV(clf,param_grid=grid_params,verbose=True)
```

```python
grid_clf.fit(x_train,y_train)
```

```python
grid_clf.best_params_
```

```python
grid_clf.score(x_test,y_test)
```

```python
grid_clf.best_estimator_
```

#### Confusion matrix with grid Search CV

```python
grid_preds = grid_clf.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test,grid_preds),cbar=False,cmap='summer',annot=True);
```

```python
print(classification_report(y_test,grid_preds))
```
