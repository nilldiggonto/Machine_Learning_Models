```python
#dropping address column as machine learning model only works with numeric data
data = data.drop('Address',axis=1)
```

#### Creating the feature and target columns

```python
featured_x = data.drop('Price',axis=1)
target_y = data['Price']
```

```python
featured_x.head()
```

### SKLEARN 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(featured_x,target_y,test_size=0.2,random_state=123)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape, y_test.shape
```

### LINEAR REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
#Finding the pattern on train data
model.fit(x_train,y_train)
```

```python
### CHecking the accuracy
model.score(x_test,y_test)
```

#### Intercept

```python
model_intercept = model.intercept_
model_intercept
```

##### Coefficient

```python
model_coef = model.coef_
model_coef
```

#### Creating a dataFrame with the intercept and coef

```python
coef_data = pd.DataFrame(model_coef,featured_x.columns,columns=['Coef'])
```

### PREDICTION

#### Linear Regression Evaluation Metrics
* Mean Absolute Error (MAE)
* Mean Squared Error (MSE)
* Root Mean Squared Error (RMSE)

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
def linear_metrics(y_test,y_preds):
    mae = mean_absolute_error(y_test,y_preds)
    mse = mean_squared_error(y_test,y_preds)
    rmse = np.sqrt(mean_squared_error(y_test,y_preds))
    
    print(f' Means Absolute Error: {mae} \n Mean Squared Error: {mse} \n Root Mean Squared Error: {rmse}')
    
```

```python
linear_metrics(y_test,y_preds)
```
